## OOV Check

```python
fitur_nostemstop = pd.read_csv("../data/data_test_nostemstop_tfidf.csv").columns[:-12]
```

```python
def generate_ngrams(s):
    
    # Break sentence in the token, remove empty tokens
    tokens = [token for token in s.str.split(" ") if (token != "" and token != " ")]
    
    # Use the zip function to help us generate n-grams
    # Concatentate the tokens into ngrams and return
    ngrams = np.concatenate(tokens)
    return ngrams

def oov(fitur,token):
    outOfVocab = []
    for i in token:
        if i not in fitur:
            outOfVocab.append(i)
    return outOfVocab

def oov_count(oovList):
    oovCount = {}
    for i in oovList:
        if i in oovCount.keys():
            oovCount[i] = oovCount[i]+1
        else:
            oovCount[i] = 1
    return oovCount
```

```python
oov_test = oov(fitur_nostemstop, generate_ngrams(svm['Tweet_Parsed']))
oov_test_word = np.array([i for i in np.unique(oov_test)if i != ''])
```

```python
def get_oov_tweet(text, oov):
    text_split = text.split(" ")
    oov_list = []
    for i in text_split:
        if i in oov:
            oov_list.append(i)    
    oov_tweet = ', '.join(map(str,oov_list))
    return oov_tweet

def putOOV(data, oov):
    data['OOV'] = data['Tweet_Parsed']
    data['OOV'] = data['OOV'].map(lambda com : get_oov_tweet(com, oov))
    return data
```

```python
svm_anl = svm.copy()
svm_anl = putOOV(svm_anl,oov_test_word)
```

```python
print("total miss classified tweet : ", len(svm))
print("miss classified tweet with oov : ",len(svm_anl[svm_anl['OOV']!=""]))
print("miss classified tweet without oov : ",len(svm_anl[svm_anl['OOV']==""]))
```

```python
pd.DataFrame({
    "condition" : ["with OOV", "without OOV"],
    "total" : [len(svm_anl[svm_anl['OOV']!=""]),len(svm_anl[svm_anl['OOV']==""])] 
}).set_index(['condition']).plot(
    kind='bar', color='red', title='jumlah oov pada miss classified tweet')
```

## Translate Check

```python
def getLang(data):
    list_lang =[]
    for i in tqdm(data['Tweet_Parsed']):
        translator = googletrans.Translator()
        lang = translator.translate(i).src
        list_lang.append(lang)
        time.sleep(2)
    return list_lang

def putLang(data, list_lang):
    data['Language'] = list_lang
    return data
```
